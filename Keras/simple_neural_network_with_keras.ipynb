```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from keras.optimizers import sgd
```

```python
model = Sequential()

layer1 = Dense(4, activation='relu', input_shape=(1,))
model.add(layer1)

#layer2 = Dense(4, activation='relu')
#model.add(layer2)

output = Dense(1, activation='linear')
model.add(output)

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
X = [[1], 
     [2], 
     [3], 
     [4]
    ]
y = [[2], 
     [4], 
     [6], 
     [8]
    ]

X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)

model.fit(X, y, epochs=5000, batch_size=4, verbose=1)
```

```python
p = model.predict([5])
print(p)

p = model.predict([6])
print(p)
```
